## freq vocab

facebook_combined 데이터는 단방향 친구 데이터임 <br>
따라서 freq vocab을 만들기 위해서는 전체 빈도를 계산하기위해 쌍방향 친구 데이터로 재구성이 필요함 <br>

way2_map_func : (0, 1)이면 (0, 1), (1, 0)을 반환하여 쌍방향 친구관계를 구축하는 map 함수 <br>
way2_friends : map함수가 (자신, 친구) pair를 만들도록 설계되었으므로 groupByKey를 통해 모아줌 <br>
number_of_friends : 전체 친구 리스트에서 등장 빈도를 파악 <br>
vocab_list : 등장 빈도가 낮은 순으로 정렬하여 최종 결과인 freq_vocab_list를 제작 <br>

```python
def way2_map_func(input_data) : 
    rnum, vec = input_data
    
    tmp_list = []
    for element in vec : 
        tmp_list.append((rnum, element))
        tmp_list.append((element, rnum))
    
    return tmp_list
```

## Sorting and Cut friends list

Inverted List를 만들 때 빈도가 낮은 순으로 정렬하여 일정한 규칙에 따른 개수로 잘라 만들어야함 <br>

cut_sort_by_list : 리스트를 정렬하여 |𝑥| − |𝜎 ⋅ |𝑥|| + 1인 prefix_size만큼 잘라 반환함
> if prefix_size == len(tmp_list) : break 부분은 필요 이상의 연산을 줄이기위한 체크

```python
# sorting by another array values
# Z = [x for _,x in sorted(zip(Y,X))]
```

```python
test = [78, 152, 181, 195, 218, 273, 275, 306, 328] # 4
len(test)
```

```python
# list 값을 기준으로 sorting하는 방법이 필요
# 자동으로 prefix size |x| - |sigma - |x|| + 1로 나오도록 변경
def cut_sort_by_list(val_list, vocab_list) : 
    tmp_list = []
    len_val_list = len(val_list)
    prefix_size = math.floor(len_val_list - CFG.sigma * len_val_list + 1)
    for val in vocab_list : 
        if val in val_list : 
            tmp_list.append(val)
            if prefix_size == len(tmp_list) : 
                break
    return tmp_list
```

## Inverted map

빈도가 낮은 순으로 정렬되고 길이가 |𝑥| − |𝜎 ⋅ |𝑥|| + 1인 friends_list를 사용하여 inverted list 제작 <br>
invert_map_func : (자신, [친구 리스트])를 받아 각 친구에 대해 (친구, 자신) pair를 반환 <br>
> 따라서 groupByKey를 통해 흩어진 key-value pair를 모을 필요가 있음

inverted_friends_list에서 (1, [친구 리스트])가 없는 이유는 1번이 |𝑥| − |𝜎 ⋅ |𝑥|| + 1에서 잘렸기 때문

```python
def invert_map_func(input_data) : 
    rnum, vec = input_data
    
    tmp_list = []
    for element in vec : 
        tmp_list.append((element, rnum))
    
    return tmp_list
```

## Jaccard Similarity

cal_jaccard : python의 set 연산을 통해 연산 수행 <br>
find_similar : jaccard 연산을 수행하는 함수
> broad_friends_list는 key값이 오름차순으로 정렬되어 있으므로 index를 통해 불러올 수 있음 <br>
> key1 < key2 일 때만 비교함으로써 중복연산을 제거할 수 있음 <br>
> 중복되는 추천 친구가 발생할 수 있으므로 ((key1, key2), Jaccard_coef)형태로 key pair를 분리

tmp_result : jaccard값을 단순 계산한 임시 결과 <br>
result : threshold를 넘는 jaccard값을 필터링하여 중복을 제거하면 최종 결과가 됨 <br>

```python
def cal_jaccard(vec1, vec2) : 
    set1 = set(vec1)
    set2 = set(vec2)
    return len(set1.intersection(set2)) / len(set1.union(set2))
```

```python
print(broad_friends_list.value[4])
print(broad_friends_list.value[5])
print(cal_jaccard(broad_friends_list.value[4][1], broad_friends_list.value[5][1]))
```

```python
def find_similar(input_data) : 
    friends_list = [broad_friends_list.value[num] for num in input_data[1]]
    tmp_list = []
    for vec1 in friends_list : 
        for vec2 in friends_list : 
            if (vec1[0] < vec2[0]) : 
                jaccard_coef = cal_jaccard(vec1[1], vec2[1])
                tmp_list.append(((vec1[0], vec2[0]), jaccard_coef))
    return tmp_list
```

```python
tmp_result = inverted_friends_list.flatMap(lambda x : find_similar(x))
```

```python
tmp_result = tmp_result.sortBy(lambda x : x[0])
```
